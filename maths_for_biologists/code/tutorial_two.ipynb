```python
from ipywidgets import interact
import matplotlib.pyplot as plt
import numpy as np
```

```python
def func(x):
    return x ** 3
```

```python
def htrans(c):
    x = np.linspace(-3,3)
    y = func(x+c)
    
    plt.plot(x, y, label = "$f(x+c)$")
    plt.title("Horizontal translation")
    plt.legend(fontsize = 12)
    
    plt.ylim(-8,8)
    
    plt.axhline(0, color = "black", lw = 1)
    plt.axvline(0, color = "black", lw = 1)
    
    plt.show()
```

```python
def vtrans(c):
    x = np.linspace(-3,3)
    
    y = func(x) + c
    
    plt.plot(x, y, label = "$f(x)+c$")
    plt.title("Vertical translation")
    
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend(fontsize = 12)
    
    plt.ylim(-8, 8)
    
    plt.axhline(0, color = "black", lw = 1)
    plt.axvline(0, color = "black", lw = 1)
    
    plt.show()
```

```python
interact(htrans, c = (-2.0, 2.0));
interact(vtrans, c = (-2.0, 2.0))
```

```python
def x(t, w = np.pi / 4, phi = 0):
    ans = np.round(np.cos(np.multiply(t, w) + phi), 2)
    return ans

def y(t, w = np.pi / 4, phi = 0):
    ans = np.round(np.sin(np.multiply(t, w) + phi), 2)
    return ans
```

```python
t = np.linspace(0,8,100)

# print(t)
# print(x(t))
# print(t)
# print(y(t))

fig, axs = plt.subplots(2, 2, gridspec_kw={'width_ratios': [3, 1]})
fig.set_figheight(5)
fig.set_figwidth(10)

# axs[0,1].figure(figsize = (5,5), dpi = 100)
axs[0,1].plot(x(t), y(t))
axs[0,0].plot(t, y(t))
axs[1,1].plot(x(t),t)
axs[1,1].invert_yaxis()
axs[1,0].grid(False)
axs[1,0].axis("off")
```

```python
chi = - np.pi / 4

fig, axs = plt.subplots(2, 2, gridspec_kw={'width_ratios': [3, 1]})
fig.set_figheight(5)
fig.set_figwidth(10)

# axs[0,1].figure(figsize = (5,5), dpi = 100)
axs[0,1].plot(x(t, chi), y(t, chi))
axs[0,0].plot(t, y(t, chi))
axs[1,1].plot(x(t, chi),t)
axs[1,1].invert_yaxis()
axs[1,0].grid(False)
axs[1,0].axis("off")
```
