```python
rN_1 = []
rN_2 = []
rN_3 = []

for n in range(0,21):
    rN_1.append((5 * n) / (1 + n))
    rN_2.append((5 * n) / (3 + n))
    rN_3.append((8 * n) / (1 + n))
```

```python
import matplotlib.pyplot as plt
plt.plot(rN_1)
plt.plot(rN_2)
plt.plot(rN_3)
plt.legend(["a = 5, k = 1", "a = 5, k = 3", "a = 8, k = 1"])
plt.draw()
plt.show()
```

```python
fx_1 = []
fx_2 = []
fx_3 = []

for x in range(0,21):
    fx_1.append((x ** 1) / ((2 ** 1) + (x ** 1)))
    fx_2.append((x ** 2) / ((2 ** 2) + (x ** 2)))
    fx_3.append((x ** 3) / ((2 ** 3) + (x ** 3)))
```

```python
plt.plot(fx_1)
plt.plot(fx_2)
plt.plot(fx_3)
plt.legend(["n = 1", "n = 2", "n = 3"])
plt.draw()
plt.savefig("results/tut1_q4.png")
plt.show()
```

```python
import csv
with open("data/words_moby_dick.csv", "r") as f:
    words = csv.reader(f)
    header = next(words)
    occ = []
    label = []
    for row in words:
        occ.append(int(row[1]))
        label.append(int(row[2]))
```

```python
import numpy as np

# label_array = np.asarray(label)
# print(np.log(label_array[0]))
# label_log = np.log(label_array)
label = np.asarray(label)
occ = np.asarray(occ)
model = np.polyfit(np.log(label),np.log(occ),1)
fig, axs = plt.subplots(nrows = 2, ncols = 1)
axs[0].plot(label, occ)
axs[1].plot(np.log(label), np.log(occ))
axs[1].plot(model)
plt.draw()
plt.show()

print(model)
```
